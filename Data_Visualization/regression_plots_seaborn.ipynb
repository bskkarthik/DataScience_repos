Regression Plots :


```
import seaborn as sns
#sns.get_dataset_names()
#dir(sns)
%matplotlib inline
```

```
df = sns.load_dataset('tips')
```

```
df.head()
```

```
sns.lmplot(x="total_bill", y="tip",data=df)
```

```
sns.lmplot(x="total_bill", y="tip",data=df, hue='sex')
```

```
sns.lmplot(x="total_bill", y="tip",data=df, hue='sex', palette='coolwarm')
```

```
sns.lmplot(x="total_bill", y="tip",data=df, hue='sex', palette='coolwarm', markers=['o','v'])
```

```
sns.lmplot(x="total_bill", y="tip",data=df, hue='sex', palette='coolwarm', markers=['o','v'], scatter_kws={'s':150})
```

```
sns.lmplot(x='total_bill',y='tip',data=df,col='sex')
```

```
sns.lmplot(x="total_bill", y="tip", row="sex", col="time",data=df)
```

```
sns.lmplot(x='total_bill',y='tip',data=df,col='day',hue='sex',palette='coolwarm')
```
